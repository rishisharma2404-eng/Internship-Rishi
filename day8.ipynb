```python
import pandas as pd
df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Name': ['Aman', 'Riya', 'Karan', 'Sneha']
})
df2 = pd.DataFrame({
    'ID': [3, 4, 5, 6],
    'Marks': [85, 90, 78, 88]
})
print("DataFrame df1:")
print(df1)
print("\nDataFrame df2:")
print(df2)
inner_merge = pd.merge(df1, df2, on='ID', how='inner')
print("\n--- INNER MERGE ---")
print(inner_merge)
left_join = pd.merge(df1, df2, on='ID', how='left')
print("\n--- LEFT JOIN ---")
print(left_join)
right_join = pd.merge(df1, df2, on='ID', how='right')

print("\n--- RIGHT JOIN ---")
print(right_join)

df1_index = df1.set_index('ID')
df2_index = df2.set_index('ID')

index_join = df1_index.join(df2_index, how='inner')

print("\n--- INDEX-BASED JOIN ---")
print(index_join)

print("\nComparison:")
print("merge() joins using columns.")
print("join() joins using indexes.")

df3 = pd.DataFrame({
    'ID': [1, 1, 2, 2],
    'Department': ['HR', 'IT', 'HR', 'IT'],
    'Salary': [30000, 40000, 35000, 45000]
})

df4 = pd.DataFrame({
    'ID': [1, 2, 2, 3],
    'Department': ['HR', 'HR', 'IT', 'IT'],
    'Bonus': [5000, 6000, 7000, 8000]
})

print("\nDataFrame df3:")
print(df3)

print("\nDataFrame df4:")
print(df4)

multiple_keys_merge = pd.merge(
    df3,
    df4,
    on=['ID', 'Department'],
    how='inner'
)

print("\n--- MERGE WITH MULTIPLE KEYS ---")
print(multiple_keys_merge)
```

```python
import pandas as pd

df1 = pd.DataFrame({
    'ID': [1, 2],
    'Name': ['Aman', 'Riya']
})

df2 = pd.DataFrame({
    'ID': [3, 4],
    'Name': ['Karan', 'Sneha']
})

df3 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Marks': [85, 90, 78, 88]
})

print("DataFrame df1:")
print(df1)

print("\nDataFrame df2:")
print(df2)

print("\nDataFrame df3:")
print(df3)

concat_df = pd.concat([df1, df2], axis=0)

print(" AFTER VERTICAL CONCATENATION")
print(concat_df)

merged_df = pd.merge(concat_df, df3, on='ID', how='inner')

print("MERGED DATAFRAME")
print(merged_df)

concat_index = concat_df.set_index('ID')
df3_index = df3.set_index('ID')

joined_df = concat_index.join(df3_index)

print("\n--- USING join() ---")
print(joined_df)
print("DIFFERENCE BETWEEN join() AND merge()")
print("""
1. merge()
   - Used to combine DataFrames using columns.
   - More flexible.
   - Similar to SQL joins.
   - Requires 'on' parameter.

2. join()
   - Used mainly for index-based joining.
   - Simpler and faster for index operations.
   - Joins using index by default.

3. merge() supports:
   - inner
   - left
   - right
   - outer joins

4. join() is best when:
   - indexes are already meaningful.
""")
```
