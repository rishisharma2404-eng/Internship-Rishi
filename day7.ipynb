```python
import pandas as pd
data = [
    {'Name': 'Grace', 'Age': 27, 'Occupation': 'Developer'},
    {'Name': 'Heidi', 'Age': 32, 'Occupation': 'Analyst'},
    {'Name': 'Ivan',  'Age': 45} 
]
df = pd.DataFrame(data)
print(df)
```

```python
data = [
    {'Name': 'Alice', 'Age': 25, 'Department': 'HR'},
    {'Name': 'Bob',   'Age': 30, 'Department': 'Engineering'},
    {'Name': 'Cara',  'Age': 35, 'Department': 'Engineering'},
    {'Name': 'Dave',  'Age': 40, 'Department': 'Marketing'}
]
df = pd.DataFrame(data)
for row in df.itertuples():
    print(f"Index: {row.Index}, Name: {row.Name}")
for index, row in df.iterrows():
    print(f"Index: {index}, Age: {row['Age']}")
```

```python
engineering_df = df[df['Department'] == 'Engineering']

senior_eng = df[(df['Age'] > 30) & (df['Department'] == 'Engineering')]
print(engineering_df)
print(senior_eng)
```

```python
first_row = df.iloc[0]
last_row = df.iloc[-1]
middle_rows = df.iloc[1:3]
print("First Row:\n", first_row)
print("\nLast Row:\n", last_row)
print("\nMiddle Rows:\n", middle_rows)
```

```python
limited_data = df.loc[0:2, ['Name', 'Age']]
limited_data_alt = df[['Name', 'Age']].head(3)
print(limited_data)
print(limited_data_alt)
```

```python
df = df[df['Department'] != 'Engineering']

print(df)
```

```python
new_row = pd.DataFrame([{'Name': 'Eve', 'Age': 28, 'Department': 'Sales'}])
df_inserted = pd.concat([df.iloc[:2], new_row, df.iloc[2:]]).reset_index(drop=True)
print(df_inserted)
```
