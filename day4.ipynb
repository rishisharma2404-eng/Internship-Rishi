```python
#1.
def max(a,b,c):
    if a > b and a > c:
        return a
    elif b > a and b > c:
        return b
    else:
        return c
print(max(23,5,8))
```

```python
#2.
def get_distinct(lst):
    return list(set(lst))

print(get_distinct([1, 2, 2, 3, 4, 4, 5]))
```

```python
#3.
def mul_lst(numbers):
    result = 1
    for num in numbers:
        result *= num
    return result
numbers = [ 2, 3, 4,10,5]
print(mul_lst(numbers))
```

```python
#4.
def fact(num):
    if num == 0 or num == 1:
        return 1
    else:
        return num * fact(num - 1)
num = 7
print(fact(num))
```

```python
#6.
def check_range(number, start, end):
    if number >= start and number <= end:
        return True
    else:
        return False
print(check_range(5, 1, 10))
print(check_range(15, 1, 10))
```

```python
#7.
def check_prime(num):
    if num <= -1:
        print("Negative numbers are not prime.")
    else:
        limit = int(num**0.5)
        for i in range(2,limit+1):
            if num % i == 0:
                print("number is not prime.")
                return False
        print("number is prime.")
        return True

print(check_prime(71))
print(check_prime(117))
```

```python
#9.
def count_case_letters(text):
    upper_count = 0
    lower_count = 0
    for char in text:
        if char.isupper():
            upper_count += 1
        elif char.islower():
            lower_count += 1

    print("Upper case letters:", upper_count)
    print("Lower case letters:", lower_count)
count_case_letters("Hello World")
```
