```python
import csv
with open("address_book.csv", "w", newline="") as file: #using with because it automatically closes the file after work is completed.
    #Better than manually using file.close().
    writer = csv.writer(file)
    writer.writerow(["Name", "Address", "Mobile", "Email"])
    for i in range(3):
        print(f"\nEnter details for person {i+1}")
        name = input("Enter Name: ")
        address = input("Enter Address: ")
        mobile = input("Enter Mobile Number: ")
        email = input("Enter Email: ")
        writer.writerow([name, address, mobile, email])
print("\nData successfully stored in address_book.csv")
```

```python
import sqlite3
conn = sqlite3.connect("college.db")
cursor = conn.cursor()
cursor.execute("""
CREATE TABLE IF NOT EXISTS Students (
    student_id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER,
    city TEXT
)
""")
cursor.execute("""
CREATE TABLE IF NOT EXISTS Courses (
    course_id INTEGER PRIMARY KEY,
    course_name TEXT,
    duration INTEGER
)
""")
cursor.execute("INSERT INTO Students VALUES (1, 'Rahul', 20, 'Delhi')")
cursor.execute("INSERT INTO Students VALUES (2, 'Aman', 21, 'Jaipur')")
cursor.execute("INSERT INTO Students VALUES (3, 'Riya', 19, 'Mumbai')")
cursor.execute("INSERT INTO Courses VALUES (101, 'Python', 6)")
cursor.execute("INSERT INTO Courses VALUES (102, 'Web Development', 8)")
cursor.execute("INSERT INTO Courses VALUES (103, 'Data Science', 10)")
conn.commit()
print("\nAll Students:")
cursor.execute("SELECT * FROM Students")
for row in cursor.fetchall():
    print(row)
print("\nOnly Names and Cities:")
cursor.execute("SELECT name, city FROM Students")
for row in cursor.fetchall():
    print(row)
print("\nStudents Age Greater Than 19:")
cursor.execute("SELECT * FROM Students WHERE age > 19")
for row in cursor.fetchall():
    print(row)
print("\nAll Courses:")
cursor.execute("SELECT * FROM Courses")
for row in cursor.fetchall():
    print(row)
cursor.execute("""
UPDATE Students
SET city = 'Pune'
WHERE student_id = 2
""")
conn.commit()
print("\nAfter Updating Student Data:")
cursor.execute("SELECT * FROM Students")
for row in cursor.fetchall():
    print(row)
cursor.execute("""
DELETE FROM Students
WHERE student_id = 3
""")
conn.commit()
print("\nAfter Deleting Student:")
cursor.execute("SELECT * FROM Students")
for row in cursor.fetchall():
    print(row)
conn.close()
```
